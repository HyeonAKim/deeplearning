## 데이터 구조
샘플수 : 150
속성 수 : 4
- 꽃받침 길이 (sepal lenght, 단위:cm)
- 꽃받침 넒이 (sepal width, 단위:cm)
- 꽃잎 길이(petal lenght, 단위:cm)
- 꽃잎 넓이(petal width, 단위:cm)
클래스 : iris-selosa, iris-wecicolor, iris-viginica



```python
# 데이터의 일부를 불러와 내용을 확인
import pandas as pd
df = pd.read_csv('C:\\Users\\HyunA\\PycharmProjects\\deeplearning\\dataset\\iris.csv'
                 ,names=["sepal_lenght","sepal_width","petal_length","petal_width","species"])
print(df.head())
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.pairplot(df, hue='species');
plt.show()
```

```python
# 원핫인코딩
# 데이터안에 문자열이 포함될 경우에는 numpy보다 pandas로 데이터 구분하는 것이 좋음

dataset = df.values
X = dataset[:,0:4].astype(float)
Y_obj = dataset[:,4]
```

```python
# y 문자열을 숫자로 바꿔주려면 클래스 이름을 숫자형태로 바꿔주어야함

from sklearn.preprocessing import LabelEncoder
e = LabelEncoder()
e.fit(Y_obj)
Y = e.transform(Y_obj)

print(Y)
```

```python
# 활성함수에 적용하려면 Y 값이 숫자 0과 1로 이루어져야함
from keras.utils import np_utils

Y_encoded = np_utils.to_categorical(Y)
print(Y_encoded)
```

```python
# 모델 생성하기
# 소프트 맥스 활용(출력층 활성함수) > 교차엔트로피( 오차함수) > [0,1,0] 형태로 전환 

from keras.models import Sequential
from keras.layers.core import Dense
from keras.utils import np_utils
from sklearn.preprocessing import LabelEncoder

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import  tensorflow as tf

# seed값 설정
seed = 0
np.random.seed(seed)
tf.set_random_seed(seed)

# 데이터입력
df = pd.read_csv('C:\\Users\\HyunA\\PycharmProjects\\deeplearning\\dataset\\iris.csv'
                 ,names=["sepal_lenght","sepal_width","petal_length","petal_width","species"])

# 데이터분류
dataset = df.values
X = dataset[:,0:4].astype(float)
Y_obj = dataset[:,4]

# 문자열 숫자로 변환
e = LabelEncoder()
e.fit(Y_obj)
Y = e.transform(Y_obj)
Y_encoded = np_utils.to_categorical(Y)

# 모델 설정
model = Sequential()
model.add(Dense(16,input_dim=4,activation='relu'))
model.add(Dense(3,activation='softmax'))

#모델 컴파일
model.compile(loss='categorical_crossentropy',
              optimizer='adam',
              metrics=['accuracy'])
```

```python
# 모델 실행
model.fit(X,Y_encoded, epochs=50,batch_size=1)

# 결과출력
print("\n Accuracy: %.4f" % (model.evaluate(X,Y_encoded)[1]))
```
